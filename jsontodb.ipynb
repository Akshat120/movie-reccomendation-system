```python

import pandas as pd
import sqlite3

# Reads the json generated from the CLI commands above and creates a pandas dataframe
tweets_df = pd.read_json('jsons/2023-04-16-04-33-10.json', lines=True)
```

```python
tweets_df['user'] = tweets_df['user'].apply(lambda x: ast.literal_eval(x))

# Extract the 'username' attribute into a new column called 'user_username'
tweets_df['user_username'] = tweets_df['user'].apply(lambda x: x['username'])
```

```python
tweets_df['user_id'] = tweets_df['user'].apply(lambda x: x['id'])
tweets = tweets_df[['user_id', 'date', 'renderedContent']]
```

```python
tweets
```

```python
# Convert the 'date' column to pandas Timestamp format
tweets['date'] = pd.to_datetime(tweets['date'], format='%Y-%m-%d %H:%M:%S%z')

# Define a function to convert timestamp to Unix time
def convert_to_unix_time(timestamp):
    return int(timestamp.timestamp())

# Apply the conversion function to the 'date' column and create a new column called 'unix_time'
tweets['unix_time'] = tweets['date'].apply(lambda x: convert_to_unix_time(x))
```

```python
tweets
```

```python
# Connect to the database
conn = sqlite3.connect('movie_ratings.sqlite')
cursor = conn.cursor()

# # Define the new data to be inserted
# new_data = [('Mary',), ('Alex',), ('Jamie',)]

# Define the SQL query for retrieving the current maximum ID
max_id_query = """
SELECT MAX(user_id) 
FROM users 
"""

# Execute the query to retrieve the current maximum ID
cursor.execute(max_id_query)
max_id = cursor.fetchone()[0]

# If the table is empty, set the maximum ID to 0
if max_id is None:
    max_id = 0

# Define the SQL query for inserting new data with unique IDs
insert_query = "INSERT OR IGNORE INTO users (user_id, twiter_id) VALUES (?, ?)"

# Insert each new value with a new unique ID
for value in tweets['user_id']:
    new_id = max_id + 1
    new_value = (new_id,) + (value,)
    cursor.execute(insert_query, new_value)
    max_id += 1
    
# Commit the changes to the database
conn.commit()

# Close the database connection
cursor.close()
conn.close()
```
